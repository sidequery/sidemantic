```python
import duckdb

conn = duckdb.connect(":memory:")
conn.execute("""
    create table auctions as
    select * from 'https://sampledata.sidequery.dev/sidemantic-demo/auction_data.parquet'
""")

conn.execute("SELECT * FROM auctions LIMIT 5").fetchdf()
```

```python
from sidemantic.widget import MetricsExplorer

widget = MetricsExplorer(conn.table("auctions"))
widget
```
